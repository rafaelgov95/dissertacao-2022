```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("DataSetViana/Variados/Polar_H10_AD655A21_20240917_063716_ECG.txt", sep=';')
ecg = data['ecg [uV]'].to_numpy()
ecg = ecg[1500:3000]
```

```python
plt.figure(figsize=(30,8))

plt.plot(ecg)
plt.title('Plot of ECG')  
plt.xlabel('Interval Index')
plt.ylabel('Time (milliseconds)')
plt.grid(True)
plt.show()
```
